```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

### Types of plots

#### Line Plots

#### Line Plots

```python
# Simple test plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]
plt.plot(x, y);

####### Customize the line plot ########

# Add Title
plt.title('Simple Line Plot');

# Add labels
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');

# Add grid background to the plot
plt.grid(True)
```

#### Bar Chart

```python
# Monthly revenue
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
revenue = [12000, 15000, 14000, 17000, 16000, 18000, 
           19000, 20000, 18500, 21000, 23000, 25000]

# Plot the bar chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(months, revenue, color='skyblue', edgecolor='black')

####### Customize the line plot ########

# Add titles and labels
plt.title("Monthly Revenue", fontsize=16);  # Title of the chart
plt.xlabel("Months", fontsize=12);          # Label for x-axis
plt.ylabel("Revenue ($)", fontsize=12);     # Label for y-axis

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=10);

# Add a grid for easier interpretation
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

#### Histogram

```python
# Students' marks (out of 100)
marks = [45, 67, 56, 78, 89, 34, 55, 68, 49, 92, 76, 88, 73, 65, 58, 84, 62, 79, 90, 48]
```

```python
# Create the histogram
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(marks, bins=10, color='purple', edgecolor='black', alpha=0.7);  # Histogram with 10 bins

####### Customize the Histogram ########

# Add titles and labels
plt.title("Students' Marks Distribution", fontsize=16)
plt.xlabel("Marks Range", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

#### Scatter plot

```python
# Height (in inches) and weight (in pounds)
heights = [60, 62, 64, 65, 66, 68, 70, 72, 74, 76]  # Heights in inches
weights = [110, 120, 130, 140, 145, 150, 160, 170, 180, 190]  # Weights in pounds
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(heights, weights, color='green', edgecolor='black', alpha=0.7);  # Scatter plot

####### Customize the Scatterplot ########


# Add titles and labels
plt.title("Height vs Weight", fontsize=16);
plt.xlabel("Height (inches)", fontsize=12);
plt.ylabel("Weight (pounds)", fontsize=12);


# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)
```

### Subplots and Legends

Subplots in Matplotlib allow you to create multiple plots within the same figure, and legends help identify different data series in your plots.

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

#### Subplots

```python
# First subplot - Original Price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original Price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Second subplot - Predicted Price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted Price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Third subplot - Difference between Original price and predicted price
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color='purple', label='Price Defference')
plt.title('Predicted Price over time')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
```

#### Saving a plot

```python
plt.plot(days, original_price, 'b', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');

# Save as image
# plt.savefig('example_prediction.png', dpi=300)
```
